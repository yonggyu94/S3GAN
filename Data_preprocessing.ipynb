```python
import os
from PIL import Image
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
dataset = "vgg"   # ["omniglot, vgg, celeba, animal_face, cub
k_shot = 3
```

```python
root = "/home/userC/yonggyu/data/%s_%d-shot" % (dataset, k_shot)
labeled_root = "/home/userC/yonggyu/data/%s_%d-shot/%d-shot" % (dataset, k_shot, k_shot)
unlabeled_root = "/home/userC/yonggyu/data/%s_%d-shot/%d-shot-complement" % (dataset, k_shot, k_shot)
```

```python
save_root = "/home/userB/yonggyukim/data/%s" % dataset
```

```python
for _class in os.listdir(labeled_root):
    labeled_class_root = os.path.join(labeled_root, _class)
    unlabeled_class_root = os.path.join(unlabeled_root, _class)
    save_class_root = os.path.join(save_root, _class)
    
    if not os.path.isdir(save_class_root):
        os.mkdir(save_class_root)
    
    for labeled_img in os.listdir(labeled_class_root):
        img_root = os.path.join(labeled_class_root, labeled_img)
        save_img_root = os.path.join(save_class_root, labeled_img)
        img = Image.open(img_root)
        
        img.save(save_img_root)
    
    for i, unlabled_img in enumerate(os.listdir(unlabeled_class_root)):
        if i > 26:
            break
        img_root = os.path.join(unlabeled_class_root, unlabled_img)
        save_img_root = os.path.join(save_class_root, "unlabeled_"+unlabled_img)
        img = Image.open(img_root)
        
        img.save(save_img_root)
        i = i + 1
        
    print(_class)
```

```python
save_root = "/home/userC/yonggyu/data/cub_20-shot"
if not os.path.isdir(save_root):
    os.mkdir(save_root)

for ori_root, _class in train_list1:
    _class = str(_class)
    save_class_root = os.path.join(save_root, ori_root.split("/")[-2])
    if not os.path.isdir(save_class_root):
        os.mkdir(save_class_root)
    
    img = Image.open(ori_root).convert("RGB")
    save_img_root = os.path.join(save_class_root, ori_root.split("/")[-1])
    img.save(save_img_root)
    print(ori_root.split("/")[-2])

    
for ori_root, _class in train_list2:
    save_class_root = os.path.join(save_root, ori_root.split("/")[-2])
    if not os.path.isdir(save_class_root):
        os.mkdir(save_class_root)
    
    img = Image.open(ori_root).convert("RGB")
    save_img_root = os.path.join(save_class_root, ori_root.split("/")[-1])
    img.save(save_img_root)
    print(ori_root.split("/")[-2])
```

```python
labeled_root = "/home/userC/yonggyu/data/cub_20-shot/20-shot"
save_root = "/home/userB/yonggyukim/data/%s" % "cub"

for _class in os.listdir(labeled_root):
    labeled_class_root = os.path.join(labeled_root, _class)
    save_class_root = os.path.join(save_root, _class)
    
    if not os.path.isdir(save_class_root):
        os.mkdir(save_class_root)
    
    for labeled_img in os.listdir(labeled_class_root):
        img_root = os.path.join(labeled_class_root, labeled_img)
        save_img_root = os.path.join(save_class_root, labeled_img)
        img = Image.open(img_root)
        
        img.save(save_img_root)
    
    i = 0
    for un_class, _ in test_list1:
        if _class == un_class.split("/")[-2] and i < 80:
            save_img_root = os.path.join(save_class_root, "unlabeled_"+un_class.split("/")[-1])
            img = Image.open(un_class)
            img.save(save_img_root)
            i += 1

    for un_class, _ in test_list2:
        if _class == un_class.split("/")[-2] and i < 80:
            save_img_root = os.path.join(save_class_root, "unlabeled_"+un_class.split("/")[-1])
            img = Image.open(un_class)
            img.save(save_img_root)
            i += 1
    
    print(save_img_root)
```
